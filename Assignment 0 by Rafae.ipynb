```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Read Image
img = cv2.imread(r'E:\Coursera or ONLINE\MS BAnDS\3rd Sem\CS 5783 Machine Learning\Week 1\Assignment 0\test.png')
```

```python
#Info about img
print(img.shape)
print(type(img))
print(img.dtype)
#Displaying img
plt.imshow(img)
plt.show()
```

```python
#Measuring distance between each pixel and RGB
for i in range(img.shape[0]):    #We have already seen it's 128
    for j in range(img.shape[1]):    #We have already seen it's 128
        pixelss = img[i,j,:]
        #Distances from 3 given points
        d1 = np.linalg.norm(pixelss - [255,0, 0])
        d2 = np.linalg.norm(pixelss - [0,255, 0])
        d3 = np.linalg.norm(pixelss - [0,0, 255])
        
        #Getting minimum distance
        index_min = np.argmin([d1,d2,d3])  
        
        if index_min == 0:
            img[i,j,:] = [255,0,0]
        elif index_min == 1:
            img[i,j,:] = [0,255,0]
        elif index_min == 2:
            img[i,j,:] = [0,0,255]
```

```python
#Saving this output as image
plt.imshow(img)
plt.title("Pixels Converted to Closely Matched Primary Colors (RGB")
plt.savefig("output1.png")
plt.show()
```

**Comment: It shows most of the pixels were close to blue, and we'll find this as true if we look at the original image**
